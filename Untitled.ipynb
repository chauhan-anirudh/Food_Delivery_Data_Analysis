```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use("ggplot")
sns.set_palette("Set2")
```

```python
NUM_ROWS = 1000   # records
NUM_CUSTOMERS = 5000
DAYS_BACK = 120   # past days for orders

restaurants = [
    "Domino's", "Pizza Hut", "Subway", "KFC", "Burger King",
    "McDonald's", "Taco Bell", "Starbucks", "Dunkin Donuts", "Chipotle"
]

food_items = {
    "Pizza": (150, 400),
    "Burger": (80, 200),
    "Pasta": (120, 300),
    "Sandwich": (60, 150),
    "Fries": (50, 120),
    "Chicken Wings": (100, 250),
    "Coffee": (50, 120),
    "Wrap": (100, 180),
    "Salad": (90, 200),
    "Sushi": (250, 600)
}

ratings = [1, 2, 3, 4, 5]

order_ids = np.arange(1, NUM_ROWS + 1)
customer_ids = np.random.randint(1000, 1000 + NUM_CUSTOMERS, NUM_ROWS)
restaurants_col = np.random.choice(restaurants, NUM_ROWS)
food_choices = np.random.choice(list(food_items.keys()), NUM_ROWS)
prices = [round(random.uniform(*food_items[f]), 2) for f in food_choices]
quantities = np.random.randint(1, 6, NUM_ROWS)
delivery_times = np.random.randint(15, 76, NUM_ROWS)
ratings_col = np.random.choice(ratings, NUM_ROWS)

today = datetime.today()
order_dates = [(today - timedelta(days=int(x))).date()
               for x in np.random.randint(0, DAYS_BACK, NUM_ROWS)]

df = pd.DataFrame({
    "OrderID": order_ids,
    "CustomerID": customer_ids,
    "Restaurant": restaurants_col,
    "FoodItem": food_choices,
    "Quantity": quantities,
    "Price": prices,
    "DeliveryTime": delivery_times,
    "Rating": ratings_col,
    "OrderDate": order_dates
})

df.to_csv("food_delivery_dataset.csv", index=False)

print(f"✅ Dataset generated with {NUM_ROWS} rows")
display(df.head())
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="Restaurant", order=df["Restaurant"].value_counts().index)
plt.title("Number of Orders per Restaurant")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(data=df, x="FoodItem", order=df["FoodItem"].value_counts().index)
plt.title("Most Ordered Food Items")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["DeliveryTime"], bins=20, kde=True)
plt.title("Distribution of Delivery Times (minutes)")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Rating", data=df)
plt.title("Customer Ratings Distribution")
plt.show()
```

```python
revenue = df.groupby("Restaurant")["Price"].sum().sort_values(ascending=False)
plt.figure(figsize=(10,5))
sns.barplot(x=revenue.index, y=revenue.values)
plt.title("Total Revenue per Restaurant")
plt.xticks(rotation=45)
plt.ylabel("Revenue (₹)")
plt.show()
```

```python
orders_per_day = df.groupby("OrderDate").size()
plt.figure(figsize=(12,5))
orders_per_day.plot(marker="o")
plt.title("Order Volume Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Orders")
plt.grid(True)
plt.show()
```
